```python
from datascience import *
class kmeans_basic(cluster_model):
    def __init__(self,x:pd.core.frame.DataFrame,init_state:np.matrix,max_iters:int=100):
        self.n,self.r = x.shape
        self.x = x
        self.k = init_state.max()+1
        c = init_state
        for j in range(max_iters):
            self._set_means(c)
            dist = self._get_dist(x)
            (exc,c) = (c,dist.argmin(0))
            if (exc==c).all(): break
        self.wss = np.array([dist[c[i],i] for i in range(n)]).sum()
    def init_x(self) -> pd.core.frame.DataFrame: return self.x
    def n_obs(self)  -> int: return self.n
    def n_feat(self) -> int: return self.r
    def _set_means(self,c:np.matrix):
        self.means = np.array([self.x[c==i].mean() for i in range(self.k)])
    def _get_dist(self,newx:pd.core.frame.DataFrame):
        return np.array([((newx-self.means[i,:]).values**2).sum(1) for i in range(self.k)])
    def cluster(self,newx:pd.core.frame.DataFrame) -> pd.core.series.Series:
        dist = self._get_dist(newx)
        return pd.Series({'clusters':dist.argmin(0)},index=newx.index)
    def within_ss(self): return self.wss
class kmeans(kmeans_basic):
    def __init__(self,x:pd.core.frame.DataFrame,k:int,\
                 max_iters:int=100,seed:int=-1,init_starts:int=10):
        (n,r) = x.shape
        if seed>0: np.random.seed(seed)
        states = np.random.randint(k,size=(n,init_starts))
        models = [kmeans_basic(x,init_state=states[:,i],max_iters=max_iters) for i in range(init_starts)]
        diff = np.zeros((init_starts,init_starts))
        for i in range(init_starts-1):
            for j in range(i+1,init_starts):
                dmat = models[i].means.reshape(k,1,r)-models[j].means.reshape(1,k,r)
                dmat = (dmat**2).sum(2)
                diff[i,j] = dmat.min(1).sum()
                diff[i,j] = 1 if diff[i,j]>1e-6 else 0
                diff[j,i] = diff[i,j]
        i = diff.sum(1).argmin()
        self.state = states[:,i]
        super(kmeans,self).__init__(x,init_state=self.state,max_iters=max_iters)
class auto_kmeans(kmeans_basic):
    def __init__(self,x:pd.core.frame.DataFrame,max_k:int=10,\
                 max_iters:int=100,seed:int=-1,init_starts:int=10):
        models = {i:kmeans(x,k=i,max_iters=max_iters,seed=seed,\
                           init_starts=init_starts) for i in range(1,max_k)}
        self.wss = {i:models[i].within_ss() for i in range(1,max_k)}
        dss = [0]*(max_k-1)
        for i in range(1,max_k-1):
            dss[i] = self.wss[i]-self.wss[i+1]
            dss[i] = dss[i-1] if np.isnan(dss[i]) else dss[i]
        dss[0] = self.wss[1]/np.log(x.shape[0])
        self.ratio = np.array([dss[i]/dss[i+1] for i in range(max_k-2)])
        self.k = self.ratio.argmax()+1
        self.state = models[self.k].state
        super(auto_kmeans,self).__init__(x,init_state=self.state,max_iters=max_iters)

auto_kmeans(x).k
```

TREE BASED MODELS

```python
def treeproc(x,y,label=''):
    splits = {}
    RSSes = {}
    for name in x.columns:
        x1 = x[name]
        RSS = []
        for item in np.unique(x1)[:-1]:
            y1,y2 = (y[x1<=item],y[x1>item])
            v1,v2 = (y1.var()*len(y1),y2.var()*len(y2))
            v2 = 0 if np.isnan(v2) else v2
            RSS += [v1+v2]
        if not RSS: continue
        i = np.argmin(RSS)
        splits[name] = np.unique(x1)[i]
        RSSes[name] = RSS[i]
    if not RSSes: return
    name = min(RSSes)
    split = splits[name]
    RSS = RSSes[name]
    print(label,name,split,RSS)
    x1 = x[name]
    lowerx = x.loc[x1<=split]
    upperx = x.loc[x1>split]
    lowery = y.loc[x1<=split]
    uppery = y.loc[x1>split]
    if len(np.unique(x1[x1<=split]))<=1:
        lowerx = lowerx.drop(name,1)
    if len(np.unique(x1[x1>split]))<=1:
        upperx = upperx.drop(name,1)
    if not lowerx.empty:
        treeproc(lowerx,lowery,label+'L')
    if not upperx.empty:
        treeproc(upperx,uppery,label+'R')
```
